```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

df = pd.read_csv(r'C:\Users\arman\Downloads\allmovies.csv')
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
# Looks for missing DAta

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
# Data types for columns

df.dtypes
```

```python
# Change data type of the column

df['budget'].round().astype('Int64')
df['budget'] = df['budget'].astype('Int64')

df['gross'].round().astype('Int64')
df['gross'] = df['gross'].astype('Int64')

```

```python
# Create correct year column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
# Order our Data a little bit to see

df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
pd.set_option('display.max_rows' , 20)
```

```python
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Scatter plit with budget vs gross revenue

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Revenue')

plt.xlabel('Gross Revenue')

plt.ylabel('Budget for Film')
plt.show
```

```python
df.corr(numeric_only = 'True')
```

```python
correlation_matrix = df.corr(numeric_only = 'True')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```
